```python
t_enc_df = t_df.copy()
```

```python
from sklearn.preprocessing import LabelEncoder

gender_encoder = LabelEncoder()
genders = gender_encoder.fit_transform(t_df['Sex'].tolist())
t_enc_df['Sex'] = genders
```

```python
# 수정된 코드
t_enc_df['Age'] = t_enc_df['Age'].round().astype(int)
t_enc_df
```

```python
feature_df = t_enc_df.drop(columns=['Survived'])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score


features, targets = feature_df , target_df
X_train, X_test, y_train, y_test = train_test_split(features, targets, test_size=0.2, random_state=124)

dtc = DecisionTreeClassifier()

parameters = {'max_depth':[5,6,7], 'min_samples_split':[15,16,17]}
```

```python
from sklearn.model_selection import GridSearchCV
t_dtc = GridSearchCV(dtc, param_grid=parameters, cv=5, refit=True, return_train_score=True, n_jobs=-1)
```

```python
t_dtc.fit(X_train, y_train)
```

```python
t_dtc.cv_results_
```

```python
result_df = pd.DataFrame(t_dtc.cv_results_)
result_df
```
