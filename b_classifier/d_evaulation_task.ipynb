### Confusion Matrix Task

##### 😊 내 풀이
##### 가정: 고암(Abnormal) 환자와 정상(Normal) 환자를 분류하는 의료 진단 시나리오.

- 양성(Positive): 고암 환자를 의미.
- 음성(Negative): 정상 환자를 의미.

#### 문제 1: 유방암 진단

>True Positive (TP): 모델이 양성이고 예측한 값도 양성일 떄 (맞춤)

>False Positive (FP): 모델이 음성이지만 예측한 값은 양성일 때 (틀림)
>
>True Negative (TN): 모델이 음성이고 예측한 값도 음성일 떄 (맞춤)
>
>False Negative (FN): 모델이 양성이지만 예측한 값은 음성일 때 (틀림)
>

>정밀도 (Precision): 모델이 예측을 Positive로 한 대상 중, 예측 값과 실제 값이 Positive로 일치한 데이터의 비율
> 
>재현율 (Recall): 실제 고암 환자로 한 대상 중 얼마나 놓치지 않고 고암 환자로 예측 했는가의 비율 
> 재현율을 높혀야 하므로 임계치를 낮춰야 한다. (실제로 암인 사람에 대해서 얼마나 정확하게 암이라고 예측 하는지가 중요하기 때문에)

##### 가정: 정상 메일과 스팸 메일을 분류하는 스팸 메일 분류 시나리오.

- 양성(Positive): 스팸 메일을 의미.
- 음성(Negative): 정상 메일을 의미.

#### 문제 2: 스팸 메일 필터링

> True Positive (TP): 모델을 예측한 값이 스팸 메일이고 실제 값도 스팸 메일일떄 
False Positive (FP): 모델을 예측한 값이 스팸 메일이지만 실제 값은 정상 메일 일때  
True Negative (TN): 모델을 예측한 값이 정상 메일이고 실제 값도 정상 메일일때 
False Negative (FN): 모델을 예측한 값이 정상메일이지만 실제 값은 스팸 메일일때   

> 정밀도 (Precision): 예측을 스팸메일로 한 대상 중 예측 값과 실제 값이 스팸메일로 일치한 데이터의 비율  
재현율 (Recall): 실제 값이 스펨메일인 대상 중 얼마나 놓치지 않고 스팸 메일로 예측했는가의 비율

> 정밀를 높혀야 하므로, 임계치를 높혀야 한다! 스팸메일의 경우 스펨메일을 찾는게 아니라 중요한 메일을 스팸처리 하지 않는 것

##### 가정: 범죄 예측 시나리오.

- 양성(Positive): 범죄가 예측되는 경우 (예: 범죄가 발생함).
- 음성(Negative): 범죄가 예측되지 않는 경우 (예: 범죄가 발생하지 않음).

#### 문제 3: 범죄 예측

> True Positive (TP): 모델이 범죄가 예측되는 경우로 예측 하였고, 실제 값도 범죄가 예측되었을 때 
False Positive (FP): 모델이 범죄가 예측되는 경우로 예측 하였지만, 실제 값은 범죄가 예측 되지 않았을 때 
True Negative (TN): 모델이 범죄가 예측 되지 않았을 경우로 예측 하였고, 실제 값도 범죄가 예측 되지 않았을 때 
False Negative (FN): 모델이 범죄가 예측 되지 않았을 경우로 예측 하였지만, 실제 값은 범죄가 예측 되었을 때 

> 정밀도 (Precision): 범인 중 예측을 범죄가 예측되는 경우인 대상 중에서 실제 값이 범죄가 예측되는 경우와 일치하는 데이터의 비율  
재현율 (Recall): 실제 범죄 중에서 모델이 범죄가 예측되는 대상 중 얼마나 놓치지 않고 범죄가 예측 되었는지의 비율
> 재현율을 높혀야 하므로 임계치를 낮춰야 한다. (실제 범죄 상황을 놓치지 않고 검거해야 하므로) 

##### 가정: 검색 엔진에서 웹 페이지의 검색 결과를 분류하는 검색 품질 모델 서비스 시나리오.

- 양성(Positive): 검색 결과가 사용자의 의도에 부합하는 경우 (예: 정확한 검색 결과).
- 음성(Negative): 검색 결과가 사용자의 의도에 부합하지 않는 경우 (예: 부정확한 또는 불량한 검색 결과).

#### 문제 4: 검색 품질 모델

> True Positive (TP): 모델이 정확한 검색결과로 예측 하였을 때 실제 값도 정확한 검색 결과가 나왔을 때 
False Positive (FP): 모델이 정확한 검색 결과로 예측 하였지만 실제 값은 부정확한 검색 결과 일때 
True Negative (TN): 모델이 부정확한 검색 결과로 예측 하였을 떄 실제 값도 부정확한 검색 결과가 나왔을 때 
False Negative (FN): 모델이 부정확한 검색 결과로 예측 하였지만 실제 값은 정확한 검색 결과가 나왔을 때 

> 정밀도 (Precision): 모델이 정확한 검색결과로 예측되는 대상 중 해당 예측 값과 실제 값인 정확한 검색 결과와 일치하는 데이터의 비율    
재현율 (Recall): 실제로 정확한 검색 결과 중에서  얼마나 놓치지 않고 정확한 검색 결과로 예측 되었는지의 비율

>정밀도를 높혀야 하므로 임계치를 높혀야 한다. (더 정확한 검색 정보를 얻어야 하므로)
##### 가정: 보안 시스템에서 침입 감지 시스템 시나리오.

- 양성(Positive): 침입이 감지된 경우 (예: 보안 이벤트가 발생함).
- 음성(Negative): 침입이 감지되지 않은 경우 (예: 정상 상황).  

#### 문제 5: 침입 감지 시스템

> True Positive (TP): 모델이 양성인 경우에 예측 값도 양성인 경우  
False Positive (FP): 모델이 음성인 경우에 예측 값은 양성인 경우  
True Negative (TN): 모델이 음성인 경우에 예측 값도 음성인 경우 
False Negative (FN): 모델이 양성인 경우에 예측 값은 음성인 경우  

> 정밀도 (Precision): 모델이 침입이 감지되는 경우로 예측 되었을 때 해당 예측 값과 실제 값인 침입이 감지된 경우인 것과 일치하는 데이터의 비율  
재현율 (Recall): 실제 침입 중에서 모델이 실제로 침입이 감지 되었을 때 얼마나 놓치지 않고 침입이 감지된 경우로 예측 되었는지의 비율
> 재현율을 높혀야 하므로 임계치를 높야 한다. (FP=>침입이 감지된 예측이 틀렸을 가능성을 낮춰야함)
