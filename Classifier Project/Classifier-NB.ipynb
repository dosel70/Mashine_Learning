```python
import pandas as pd
```

```python
s_df = pd.read_csv('./datasets/Sports.csv')
s_df
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
targets = encoder.fit_transform(s_df.Label)
s_df['targets'] = targets
```

```python
s_df 
```

```python
s_df = s_df.drop(labels=['Label'], axis=1)
s_df
```

```python
s_df.isna().sum()
```

```python
s_df.duplicated().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = \
train_test_split(s_df.Text, s_df.targets, stratify=s_df.targets, test_size=0.2, random_state=124)
```

```python
s_df.targets.value_counts()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline

m_nb_pipe = Pipeline([('count_vectorizer', CountVectorizer()), 
                      ('multinomial_NB', MultinomialNB())])
```

```python
m_nb_pipe.fit(X_train, y_train)
```

```python
print(y_test)
print(m_nb_pipe.predict(X_test))
```

```python
m_nb_pipe.score(X_test, y_test)
```

```python
encoder.classes_[m_nb_pipe.predict(X_test)]
```

```python
import matplotlib.pyplot as plt

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

# 타겟 데이터와 예측 객체를 전달받는다.
def get_evaluation(y_test, prediction, classifier=None, X_test=None):
#     오차 행렬
    confusion = confusion_matrix(y_test, prediction)
#     정확도
    accuracy = accuracy_score(y_test , prediction)
#     정밀도
    precision = precision_score(y_test , prediction)
#     재현율
    recall = recall_score(y_test , prediction)
#     F1 score
    f1 = f1_score(y_test, prediction)
#     ROC-AUC
    roc_auc = roc_auc_score(y_test, prediction)
    
    print('오차 행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}, F1: {3:.4f}, ROC-AUC: {4:.4f}'.format(accuracy, precision, recall, f1, roc_auc))
    print("#" * 80)
    
    if classifier is not None and  X_test is not None:
        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,4))
        titles_options = [("Confusion matrix", None), ("Normalized confusion matrix", "true")]

        for (title, normalize), ax in zip(titles_options, axes.flatten()):
            disp = ConfusionMatrixDisplay.from_estimator(classifier, X_test, y_test, ax=ax, cmap=plt.cm.Blues, normalize=normalize)
            disp.ax_.set_title(title)
        plt.show()
```

```python
prediction = m_nb_pipe.predict(X_test)
```

```python
get_evaluation(y_test, prediction, m_nb_pipe, X_test)
```
